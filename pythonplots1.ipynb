## Include

```python
import matplotlib.font_manager as fm
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm

# To use cmr
from matplotlib import rc
```

The includes are a bit messy - no need, e.g., to import rc separately.

Easier approach:

plt.rcParams['font.family'] = 'Times New Roman'

or 

plt.rc('font',family='Times New Roman')

To rebuild the font table:

matplotlib.font_manager._rebuild()

## Fonts available

```python
# Collect all the font names available to matplotlib
font_names = [f.name for f in fm.fontManager.ttflist]
print(sorted(font_names, key=lambda v: v.upper()))
```

## Plot options

```python
# Data
x = np.linspace(-6, 6, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Edit the font, font size, and axes width
#mpl.rcParams['font.family'] = 'Times'
#plt.rcParams['font.size'] = 18
#plt.rcParams['axes.linewidth'] = 2

# or use cmr
# Plain TeX
#rc('text', usetex=True)
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})

# Generate 2 colors from the 'tab10' colormap
colors = cm.get_cmap('tab10', 2)

# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(7, 5))

# Add axes object to our figure that takes up entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot data
ax.plot(x, y1, color=colors(0), label='sin')
ax.plot(x, y2, color=colors(1), label='cos')

# Set the axis limits
#ax.set_xlim(370, 930)
#ax.set_ylim(-0.2, 2.2)

# Add the x and y-axis labels
ax.set_xlabel('$a$', labelpad=10)
ax.set_ylabel('Babo', labelpad=10)
# labelpad — extra padding between the tick labels and the axis label

# Add legend to plot
ax.legend(bbox_to_anchor=(1, 1), loc=1, frameon=False, fontsize=16)

# Save figure
#plt.savefig('Final_Plot.png', dpi=300, transparent=False, bbox_inches='tight')

plt.show()
```
