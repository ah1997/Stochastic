## Q3 Stochastic gradient descent

### (a)

We define the potential as a function and finds its minima using SciPy.

```python
def Upot(x):
    return 0.5*x**4-5.0*x**2+x
```

```python
grid = (-10, 10, 0.1)
x1 = optimize.fminbound(Upot, -4.0, 0.0)  # Find local min in [-4,0]
x2 = optimize.fminbound(Upot, 0.0, 4.0)  # Find local min in [0,4]
print(x1, x2)
```

```python
# Plot potential and two minima
xview = np.linspace(-3.5, 3.5, 100)
plt.plot(xview, Upot(xview))
plt.plot([x1, x2], [Upot(x1), Upot(x2)], 'ro')
plt.xlabel(r'$x$')
plt.ylabel(r'$U(x)$')
plt.show()
```

### (b)

We solve the ODE with odeint for different random initial conditions in the range $[-4,4]$, making sure to define the force for the ODE.

```python
def fd(x, t):
    dxdt = -(2.0*x**3-10.0*x+1.0)
    return dxdt
```

```python
# Parameters
tfinal = 0.5
dt = 0.01

# Random initial conditions
x0list = [8*np.random.random()-4 for i in range(20)]

# Time vector for odeint
tspan = np.linspace(0.0, tfinal, int(tfinal/dt))

# Solve for each initial condition
for x0 in x0list:
    x = odeint(fd, x0, tspan)
    plt.plot(tspan, x, 'b-')

# Plot solutions
plt.plot([tfinal, tfinal], [x1, x2], 'ro')
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')
plt.show()
```

It is clear from the shape of the potential that initial conditions on the right of the local max, located close to $0$, descend in time to the local min and stay there, while initial conditions located on the left of the local max descend to the global min and stay there. The local max thus defines two *basins of attractions*: one for the local min and one for the global min.

The final value reached after the integration in time gives an approximation of the position of the global min:

```python
print(x[-1][0],'with absolute error',np.abs(x1-x[-1][0]))
```

### (c)

We integrate the stochastic gradient descent SDE using the Euler-Maruyama scheme, similarly to the Langevin equation, but now keep the position as a function of time:

```python
tfinal = 100.0
dt = 0.01
sigma = 2.5

n = int(tfinal/dt)
sqrtdt = np.sqrt(dt)
x = np.zeros(n+1)
x[0] = 4.0

for i in range(n):
    x[i+1] = x[i]+fd(x[i], 0)*dt+sigma*sqrtdt*np.random.randn()

tview = np.linspace(0.0, tfinal, n+1)
plt.plot(tview, x)
plt.plot(tview, [x1]*len(tview), 'r--')
plt.plot(tview, [x2]*len(tview), 'r--')
plt.xlabel(r'$t$')
plt.ylabel(r'$X_t$')
plt.show()
```

We see now that, with the noise, the state $X_t$ can transition between the two wells of the potential and diffuse around the global and the local minima, shown as the dashed lines. 

### (d)

We repeat the previous simulation, but now decrease the noise parameter in time.

```python
tfinal = 10.0
dt = 0.01
alpha = 5.0

n = int(tfinal/dt)
sqrtdt = np.sqrt(dt)
x = np.zeros(n+1)
x[0] = 4.0

for i in range(n):
    t = (i+1)*dt          # Running time
    sigma = alpha/(t+1.0)  # Noise annealing
    x[i+1] = x[i]+fd(x[i], 0)*dt+sigma*sqrtdt*np.random.randn()

tview = np.linspace(0.0, tfinal, n+1)
plt.plot(tview, x)
plt.plot(tview, [x1]*len(tview), 'r--')
plt.plot(tview, [x2]*len(tview), 'r--')
plt.show()
```

Repeating the simulation many times, we can see that $X_t$ reaches the global min most of the time and concentrates there as time goes on. The final state gives an estimate of the global min:

```python
print(x[-1], 'with absolute error', np.abs(x[-1]-x1))
```

### (e)

Stochastic gradient descent (SGD) has the advantage, over its deterministic counterpart, that it does not get stuck into local minima or, at least, gets stuck less often. Because of the noise, SGD explores more of the full space and can get out of local min that are traps for the deterministic dynamics.

SGD on its own, however, is not precise because it always wanders around a minimum when it reaches one. This is why we use annealing: we keep some noise to begin with to enable exploration and avoid getting trapped, and eventually decrease the noise in the hope that the diffusion has reached a global minimum, which is then located more precisely by letting $\sigma\rightarrow 0$.

The way the noise is decreased in time is called the *annealing schedule*. Finding the right annealing schedule for a particular optimization problem is difficult, especially in high dimensions.

## Q4 Kapitza pendulum

### (a)

We use odeint by transforming the 2nd order ODE into a 1st order system in the usual way. Define $x_0=\theta$, $x_1=\dot\theta$, and $\mathbf{x}=(x_0\ x_1)^T$. Then 
$$
\dot{\mathbf{x}}=\mathbf{f}(\mathbf{x},t),
$$
where
$$
\mathbf{f}=
\left(
\begin{array}{c}
x_1\\
-[1+A\cos(\omega t)]\sin x_0-k x_1
\end{array}
\right).
$$
Note that the force depends explicitly on time.

```python
def f_kapitza(x, t):
    A = 20.0
    omega = 10.0
    k = 0.5

    dx1dt = x[1]
    dx2dt = -(1+A*np.cos(omega*t))*np.sin(x[0])-k*x[1]

    return [dx1dt, dx2dt]
```

```python
tfinal = 20.0
dt = 0.05
tspan = np.linspace(0.0, tfinal, int(tfinal/dt))

x0 = [3*np.pi/4, 0.0]
x = odeint(f_kapitza, x0, tspan)

plt.figure(figsize=(10, 5))
plt.plot(tspan, [np.pi]*len(tspan), 'k--')
plt.plot(tspan, x[:, 0], label='Angle')
plt.plot(tspan, x[:, 1], label='Angular velocity')
plt.legend(loc='lower right')
plt.xlabel(r'$t$')
plt.ylabel(r'$\theta(t),\dot\theta(t)$')
plt.show()
```

Trying different initial conditions, we can see that the angle $\theta(t)$ converges to $\pi$ and so the upright position is stable (at least locally). For low driving frequencies, such as $\omega=1$, the upright position cannot be stabilized.

### (b)

We integrate the SDE using the Euler-Maruyama scheme.

```python
# Parameters
tfinal = 50.0
dt = 0.05
A = 20.0
omega = 10.0
k = 0.5
etalist = [0.001, 0.005, 0.01, 0.05]

n = int(tfinal/dt)
tview = np.linspace(0, tfinal, n+1)
theta = np.zeros(n+1)
gamma = np.zeros(n+1)

# Define figure space
plt.figure(figsize=(10, 5))

# Try different noise amplitudes
for eta in etalist:

    # Initial conditions
    theta[0] = 3.0*np.pi/4.0
    gamma[0] = 0.0

    # Simulation with Euler-Maruyama scheme
    for i in range(n):
        t = (i+1)*dt
        theta[i+1] = theta[i]+gamma[i]*dt
        gamma[i+1] = gamma[i]\
            - (1+A*np.cos(omega*t))*np.sin(theta[i])*dt\
            - k*gamma[i]*dt\
            + np.sqrt(eta*dt)*np.random.randn()

    # Plot trajectory
    plt.plot(tview, theta, label=r'$\eta=$'+str(eta))

# Prettify plot
plt.plot(tview, [np.pi]*len(tview), 'k--')
plt.legend(loc='lower right')
plt.xlabel(r'$t$')
plt.ylabel(r'$\theta_t$')
plt.show()
```

Amazingly, even with noise, the position still stabilizes around the upright position $\theta=\pi$, provided the noise amplitude $\eta$ is not too large.

### (c)

Trying different noise amplitudes, it seems that the threshold is roughly in the range $\eta\in [0.3,0.5]$.

To find a more precise estimate, we repeat the simulation above many times and plot the angle reached at some point in time, here $T=30$. The scatter plot of these points should tell us whether the pendulum is stable at $\pi$ or not.

```python
ntraj = 10**2
tfinal = 50.0
dt = 0.05
A = 20.0
omega = 10.0
k = 0.5
eta_range = np.arange(0., 0.2, 0.01)

n = int(tfinal/dt)
theta_mean_list = []
theta_mean_err_list = []

for eta in eta_range:

    theta_sample = []

    for j in range(ntraj):
        theta = np.pi
        gamma = 0.0

        for i in range(n):
            t = (i+1)*dt
            theta += gamma*dt
            gamma +=  - (1+A*np.cos(omega*t))*np.sin(theta)*dt\
            - k*gamma*dt\
            + np.sqrt(eta*dt)*np.random.randn()
                
        theta_sample.append(theta)

    plt.plot([eta]*len(theta_sample), theta_sample, 'b.', alpha=0.1)

plt.plot(eta_range, [np.pi]*len(eta_range), 'k--')
plt.plot(eta_range, [2*np.pi]*len(eta_range), 'r--')
plt.plot(eta_range, [0]*len(eta_range), 'r--')
plt.ylim([0-2, 2*np.pi+2])
plt.xlabel(r'$\eta$')
plt.ylabel(r'$\theta$')
plt.show()
```

The results show that angles around $\pi$ (black dashed line) become less frequent around $\eta=0.1$. At that noise power, only the lower position $\theta=0$ (or, equivalently, $\theta=2\pi$) is stable (red dashed line).

Note that, because $\theta=0$ and $\theta=2\pi$ correspond to the same position, the average of the angle is more or less always $\pi$, so the average itself cannot be used to pinpoint the destabilization transition.
